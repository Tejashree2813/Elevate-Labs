```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
df.describe()
```

```python
# Generate summary statistics
display(df.describe(include='all'))
```

Summary statistics provide an overview of the central tendency, dispersion, and shape of the dataset's distribution, as well as counts of missing values and unique values for non-numeric data.

```python
# Create histograms for numeric features
df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

Histograms show the distribution of individual numeric features, helping to identify their shape, central tendency, and potential outliers.

```python
# Create boxplots for numeric features
numeric_cols = df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

Boxplots visualize the distribution of numeric features and are particularly useful for identifying outliers and comparing distributions across different groups.

```python
# Use correlation matrix for feature relationships
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

A correlation matrix shows the pairwise correlation coefficients between numeric features, helping to understand the relationships and dependencies between them.
